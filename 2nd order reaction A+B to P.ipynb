```python
k = 0.025 #L/mol.s
C0_c2h5oac=0.5 #mol/L
C0_oh = 0.34 #mol/L

from chempy.kinetics.integrated import binary_irrev
def Ct_prod(t):
    return binary_irrev(t=t, kf=k, prod=0,major=C0_c2h5oac, minor=C0_oh)
def Ct_oh(t):
    return C0_oh-C_prod(t)
def Ct_c2h5oac(t):
    return C0_c2h5oac-C_prod(t)

print(f"Concentration of AcO- at time t = 200s is {C_prod(200):.4f} mol/L")
print(f"Concentration of NaOH at time t = 200s is {Ct_oh(200):.4f} mol/L")
print(f"Concentration of NaOH at time t = 200s is {Ct_c2h5oac(200):.4f} mol/L")
```

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,1000,1000)
c2h5oac=[Ct_c2h5oac(t_i) for t_i in t]
oh=[Ct_oh(t_i) for t_i in t]
prod=[Ct_prod(t_i) for t_i in t]

plt.plot(t,c2h5oac, label='C2H5OAc')
plt.plot(t,oh,label='NaOH')
plt.plot(t,prod, label='CH3COO- and C2H5OH')
plt.legend()
plt.title('Ester Hydrolysis')
plt.ylabel('Concentration (M)')
plt.grid(True)
plt.xlabel('Time (s)');
```
