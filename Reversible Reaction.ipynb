```python
k1=0.2 #s^-1
k_1=0.1 #s^-1
c0= {'G6P':1,'F6P':0}

from chempy import Equilibrium, ReactionSystem
from chempy.kinetics.ode import get_odesys
import numpy as np
import matplotlib.pyplot as plt
eq = Equilibrium({'G6P'},{'F6P'}, param=k1/k_1)
substances='G6P F6P'.split()
rsys=ReactionSystem(eq.as_reactions(kf=k1),substances)

odesys,extra = get_odesys(rsys)
t=np.linspace(0,40,200)
result = odesys.integrate(t,c0)

plt.plot(t, result.yout[:,0], label='[G6P]')
plt.plot(t, result.yout[:,1], label='[F6P]')
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Concentration (M)')
plt.title('Reversible Reaction')
plt.text(10, 0.95, r'$\frac{d[G6P]}{dt} = -k_1 [G6P] + k_{-1} [F6P]$', bbox=dict(facecolor='white', alpha=0.8))
plt.text(10, 0.8, r'$\frac{d[F6P]}{dt} = k_1 [G6P] - k_{-1} [F6P]$', bbox=dict(facecolor='white', alpha=0.8))

plt.legend();
```
