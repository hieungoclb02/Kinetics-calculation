```python
import numpy as np
import matplotlib.pyplot as plt

temp =np.array([300,350,400]) + 273.15
k_list = np.array([1.2e-3, 2.5e-2, 0.15])
lnk=[np.log(k) for k in k_list]
R=8.314
def Arrhenius(T, Ea, A):
    return np.log(A) - Ea/(R*T)
from scipy.optimize import curve_fit
popt, pcov = curve_fit(Arrhenius, temp, lnk, p0=(50000,1e10))
Ea, A = popt
```

```python
print(f'Ea = {Ea:.2f}(J/mol)\nA = {A:.2e} s^-1')
```
