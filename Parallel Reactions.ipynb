```python
C0=1 #mol/L
from chempy import Reaction, ReactionSystem
import numpy as np
import matplotlib.pyplot as plt
reactions=[Reaction.from_string('A -> B','A B', param=0.1),
          Reaction.from_string('A -> C','A C', param =0.05)]
rsys=ReactionSystem(reactions, ['A','B','C'])

c0={'A':1, 'B':0,'C':0}
t=np.linspace(0,70,300)

from chempy.kinetics.ode import get_odesys
odesys, extra = get_odesys(rsys)
result = odesys.integrate(t,c0)

plt.plot(t, result.yout[:,0],label='[A]')
plt.plot(t, result.yout[:,1],label='[B]')
plt.plot(t, result.yout[:,2],label='[C]')
plt.ylabel('Concentration (M)')
plt.xlabel('Time (s)')
plt.title('Parallel Reactions')
plt.grid(True)
plt.legend();
```
